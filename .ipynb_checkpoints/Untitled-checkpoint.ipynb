```python
def compute_match_weights(comparison_vectors, m_probs, u_probs, lambda_prior):
    """
    Compute P(M|γ) for each comparison vector
    """
    weights = []
    for gamma in comparison_vectors:
        # P(γ|M)P(M) / (P(γ|M)P(M) + P(γ|U)P(U))
        numerator = m_probs[gamma] * lambda_prior
        denominator = (m_probs[gamma] * lambda_prior) + (u_probs[gamma] * (1 - lambda_prior))
        weights.append(numerator / denominator)
    return np.array(weights)
```

```python
def em_algorithm(comparison_vectors, max_iter=10, tol=1e-4):
   # Initialize
   m_probs = {0: 0.02, 1: 0.90, 2: 0.98}
   u_probs = {0: 0.98, 1: 0.01, 2: 0.01} 
   lambda_prev = 0.1

   for iteration in range(max_iter):
       # E-step
       weights = compute_match_weights(comparison_vectors, m_probs, u_probs, lambda_prev)
       
       # M-step
       # Calculate lambda using pattern frequencies
       pattern_counts = {}
       for gamma in comparison_vectors:
           pattern_counts[gamma] = pattern_counts.get(gamma, 0) + 1
           
       lambda_new = 0
       for gamma, count in pattern_counts.items():
           lambda_new += (count * np.mean(weights[comparison_vectors == gamma]))
       lambda_new = lambda_new / len(comparison_vectors)
       
       # Update m and u probs considering pattern frequencies
       for level in [0, 1, 2]:
           level_indices = comparison_vectors == level
           if np.any(level_indices):
               count = pattern_counts.get(level, 0)
               m_probs[level] = (count * np.sum(weights[level_indices])) / (len(comparison_vectors) * np.sum(weights))
               u_probs[level] = (count * np.sum((1-weights)[level_indices])) / (len(comparison_vectors) * np.sum(1-weights))

       if abs(lambda_new - lambda_prev) < tol:
           break
           
       lambda_prev = lambda_new

   return weights, m_probs, u_probs, lambda_new
```

```python
# Let's run the complete test
vectors, pairs, sims = compute_string_comparisons(dataset_a, dataset_b)
weights, m_probs, u_probs, lambda_final = em_algorithm(vectors)

print(f"Number of comparisons: {len(vectors)}")
print("\nFrequency of comparison levels:")
for level in [0,1,2]:
   print(f"Level {level}: {np.sum(vectors == level)}")

print(f"\nFinal parameters:")
print(f"m probabilities: {m_probs}")
print(f"u probabilities: {u_probs}")
print(f"lambda: {lambda_final}")

# Sample some high probability matches
high_prob_indices = np.where(weights > 0.9)[0]
print("\nSample high probability matches:")
for idx in high_prob_indices[:5]:
   print(f"{pairs[idx]} - Weight: {weights[idx]:.3f}")
```
